# tool_01 auto: 在母目录创建 absorption 结构（example）

## 用法
- 你只需要指定 `parent_root`
- 脚本会在该母目录下创建标准 `absorption` 结构
- 本 notebook 只做目录建立与说明文件写入，不做任何计算

```python
from pathlib import Path
```

```python
# ---------- 用户参数（先改这里）----------
# 母目录（你后续告诉我具体路径）
parent_root = Path('/home/ucaqzh0/thermol/EXAMPLE_PARENT')

# absorption 目录名（默认 absorption）
absorption_folder_name = 'absorption'

# 是否写入 README 说明文件
write_readme_files = True

# 若目标已存在：True=保留已有内容继续补齐，False=直接报错
allow_existing = True
```

```python
def build_absorption_tree(parent: Path, folder_name: str = 'absorption', write_readme: bool = True, allow_existing_root: bool = True):
    root = parent / folder_name

    if root.exists() and not allow_existing_root:
        raise FileExistsError(f'目标目录已存在: {root}')

    # 固定接口
    bodies = ['1_body', '2_body']
    stages = ['GO', 'TEST', 'Frequency', 'Final']

    created = []
    root.mkdir(parents=True, exist_ok=True)
    created.append(root)

    for body in bodies:
        body_dir = root / body
        body_dir.mkdir(parents=True, exist_ok=True)
        created.append(body_dir)

        for stage in stages:
            d = body_dir / stage
            d.mkdir(parents=True, exist_ok=True)
            created.append(d)

    if write_readme:
        (root / 'README.md').write_text(
            '# absorption\n\n'
            '标准接口：`1_body/2_body/{GO,TEST,Frequency,Final}`\n',
            encoding='utf-8'
        )

        for body in bodies:
            (root / body / 'README.md').write_text(
                f'# {body}\n\n'
                '- GO: 全局优化主计算目录\n'
                '- TEST: 试验结构目录\n'
                '- Frequency: 虚频修正/一次SCF中间目录\n'
                '- Final: 最终确认结构目录\n',
                encoding='utf-8'
            )
            for stage in stages:
                (root / body / stage / 'README.txt').write_text(
                    f'{body}/{stage}: 目录说明占位文件。\n',
                    encoding='utf-8'
                )

    return root, created
```

```python
root, created_paths = build_absorption_tree(
    parent=parent_root,
    folder_name=absorption_folder_name,
    write_readme=write_readme_files,
    allow_existing_root=allow_existing,
)

print('✅ absorption 结构建立完成')
print(f'- root: {root}')
print(f'- created_or_checked: {len(created_paths)} paths')
for p in created_paths:
    print('  -', p)
```

## 输出说明
- 该 notebook 只负责“目录结构搭建”
- 目录结构固定为：`absorption/1_body|2_body/{GO,TEST,Frequency,Final}`
- 你后续只需要改 `parent_root`，即可在任意母目录重建同样接口

```python
# 可选：快速检查关键目录是否存在
expected = [
    root / '1_body' / 'GO',
    root / '1_body' / 'TEST',
    root / '1_body' / 'Frequency',
    root / '1_body' / 'Final',
    root / '2_body' / 'GO',
    root / '2_body' / 'TEST',
    root / '2_body' / 'Frequency',
    root / '2_body' / 'Final',
]

missing = [p for p in expected if not p.exists()]
if not missing:
    print('✅ 关键目录检查通过')
else:
    print('❌ 缺失目录:')
    for p in missing:
        print(' -', p)
```
